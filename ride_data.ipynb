```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/waiting_times.csv")
```

```python
df
```

```python
df["WORK_DATE"] = pd.to_datetime(df["WORK_DATE"])
```

```python
df = df.sort_values(by="WORK_DATE", ascending=False)
```

```python
# Extract the first non-zero capacity for each ride
def first_non_zero(series):
    # Find the first non-zero value in the series
    non_zero_values = series[series != 0]
    if not non_zero_values.empty:
        return non_zero_values.iloc[0]
    else:
        return 0  # Return 0 if all values are zero
```

```python
# Extract ride capacity for each ride by taking the first non-zero value for each ride
ride_capacity = df.groupby("ENTITY_DESCRIPTION_SHORT")["CAPACITY"].agg(first_non_zero)
```

```python
ride_capacity = ride_capacity.reset_index()
```

```python
ride_capacity.rename(columns={"ENTITY_DESCRIPTION_SHORT": "RIDE"}, inplace=True)
```

```python
ride_capacity
```

```python
ride_capacity.to_csv("data/ride_capacity.csv", index=False)
```
