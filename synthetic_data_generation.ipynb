```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/attendance.csv")

df.head()
```

```python
# Find average atttendance in PortAventura World in May 2022 on Weekends

df["USAGE_DATE"] = pd.to_datetime(df["USAGE_DATE"])

df["month"] = df["USAGE_DATE"].dt.month
df["year"] = df["USAGE_DATE"].dt.year
df["day"] = df["USAGE_DATE"].dt.day

df["day_of_week"] = df["USAGE_DATE"].dt.day_name()
```

```python
df[
    (df["FACILITY_NAME"] == "PortAventura World")
    & (df["month"] == 5)
    & (df["year"] == 2022)
    & (df["day_of_week"].isin(["Saturday", "Sunday"]))
]["attendance"].mean()
```

```python
ride_mapping = pd.read_csv("data/link_attraction_park.csv", sep=";")

ride_mapping
```

```python
portaventura_rides = ride_mapping[ride_mapping["PARK"] == "PortAventura World"]
portaventura_rides = portaventura_rides["ATTRACTION"].to_list()
portaventura_rides
```

```python
waiting_times = pd.read_csv("data/waiting_times.csv")

waiting_times
```

```python
waiting_times["ENTITY_DESCRIPTION_SHORT"].unique()
```

```python
waiting_times["WORK_DATE"] = pd.to_datetime(waiting_times["WORK_DATE"])
```

```python
# Filter out weekends in May 2022

waiting_times["month"] = waiting_times["WORK_DATE"].dt.month
waiting_times["year"] = waiting_times["WORK_DATE"].dt.year
waiting_times["day"] = waiting_times["WORK_DATE"].dt.day

waiting_times["day_of_week"] = waiting_times["WORK_DATE"].dt.day_name()
```

```python
# Find the busiest rides by DEB_TIME_HOUR

waiting_times["DEB_TIME_HOUR"] = waiting_times["DEB_TIME_HOUR"].astype(int)

waiting_times = waiting_times[
    (waiting_times["ENTITY_DESCRIPTION_SHORT"].isin(portaventura_rides))
    & (waiting_times["month"] == 5)
    & (waiting_times["year"] == 2022)
    & (waiting_times["day_of_week"].isin(["Saturday", "Sunday"]))
]
```

```python
# Filter out debt_time_hour between 10 and 19

waiting_times = waiting_times[
    (waiting_times["DEB_TIME_HOUR"] >= 10) & (waiting_times["DEB_TIME_HOUR"] <= 19)
]
```

```python
waiting_times
```

```python
# Plot the busiest rides by DEB_TIME_HOUR

plt.figure(figsize=(12, 16))
ax = sns.lineplot(
    data=waiting_times,
    x="DEB_TIME_HOUR",
    y="WAIT_TIME_MAX",
    hue="ENTITY_DESCRIPTION_SHORT",
    palette=sns.color_palette(
        "husl", n_colors=len(waiting_times["ENTITY_DESCRIPTION_SHORT"].unique())
    ),
)

# Add numeric labels to the lines
for line, name in zip(ax.lines, waiting_times["ENTITY_DESCRIPTION_SHORT"].unique()):
    y = line.get_ydata()[-1]
    x = line.get_xdata()[-1]
    if not np.isfinite(y):
        y = next(reversed(line.get_ydata()[~line.get_ydata().mask]), float("nan"))
    if not np.isfinite(y):
        continue
    ax.text(x, y, name, color=line.get_color(), fontsize=8, alpha=0.7)

plt.title("Busiest rides by WAITING TIME MAX in PortAventura World in May 2022")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
# Plot the busiest rides by GUEST_CARRIED

plt.figure(figsize=(12, 8))
ax = sns.lineplot(
    data=waiting_times,
    x="DEB_TIME_HOUR",
    y="GUEST_CARRIED",
    hue="ENTITY_DESCRIPTION_SHORT",
    palette=sns.color_palette(
        "husl", n_colors=len(waiting_times["ENTITY_DESCRIPTION_SHORT"].unique())
    ),
)

# Add numeric labels to the lines
for line, name in zip(ax.lines, waiting_times["ENTITY_DESCRIPTION_SHORT"].unique()):
    y = line.get_ydata()[-1]
    x = line.get_xdata()[-1]
    if not np.isfinite(y):
        y = next(reversed(line.get_ydata()[~line.get_ydata().mask]), float("nan"))
    if not np.isfinite(y):
        continue
    ax.text(x, y, name, color=line.get_color(), fontsize=8, alpha=0.7)

plt.title("Busiest rides by GUEST_CARRIED in PortAventura World in May 2022")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
# Add a column for capacity utilization

waiting_times["capacity_utilization"] = (
    waiting_times["GUEST_CARRIED"] / waiting_times["ADJUST_CAPACITY"]
)
```

```python
## Plot the busiest rides by capacity

plt.figure(figsize=(20, 10))  # Increase figure size
ax = sns.lineplot(
    data=waiting_times,
    x="DEB_TIME_HOUR",
    y="capacity_utilization",
    hue="ENTITY_DESCRIPTION_SHORT",
    palette=sns.color_palette(
        "husl", n_colors=len(waiting_times["ENTITY_DESCRIPTION_SHORT"].unique())
    ),
)

# Add numeric labels to the lines
for line, name in zip(ax.lines, waiting_times["ENTITY_DESCRIPTION_SHORT"].unique()):
    y_data = line.get_ydata()
    x_data = line.get_xdata()
    if len(y_data) > 0 and len(x_data) > 0:
        y = y_data[-1]
        x = x_data[-1]
        if not np.isfinite(y):
            y = next(reversed(y_data[~np.isnan(y_data)]), float("nan"))
        if not np.isfinite(y):
            continue
        ax.text(x, y, name, color=line.get_color(), fontsize=8, alpha=0.7)

plt.title("Busiest rides by capacity utilization (%) in PortAventura World in May 2022")
plt.legend(
    bbox_to_anchor=(1.02, 1), loc="upper left", borderaxespad=0.0
)  # Adjust legend position
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to make space for the legend
plt.show()
```

```python
# Most popular rides by GUEST_CARRIED

popularity = (
    waiting_times.groupby("ENTITY_DESCRIPTION_SHORT")["GUEST_CARRIED"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)
popularity
```

```python
popularity["rank"] = popularity["GUEST_CARRIED"].rank(ascending=False)
```

```python
popularity.drop("GUEST_CARRIED", axis=1, inplace=True)
popularity
```

```python
num_people = 50000
entry_times = np.random.randint(10, 15, size=num_people)
exit_times = np.random.randint(entry_times + 4, 19, size=num_people)

# Generate ride preferences based on popularity ranks
# Weights are inversely proportional to rank
weights = 1 / popularity["rank"]
weights /= weights.sum()
```

```python
# Assign three ride preferences for each person
ride_preferences = np.random.choice(
    popularity["ENTITY_DESCRIPTION_SHORT"], size=(num_people, 3), p=weights
)
```

```python
# Create the DataFrame
user_data = pd.DataFrame(
    {
        "entry_time": entry_times,
        "exit_time": exit_times,
        "ride_preference_1": ride_preferences[:, 0],
        "ride_preference_2": ride_preferences[:, 1],
        "ride_preference_3": ride_preferences[:, 2],
    }
)
```
