#### 1.6.3 Waiting with Weather

```python
def merge_nearest_weather(waiting_times_df, weather_df):
    """
    Merge weather data to waiting times based on closest timestamp match

    Parameters:
    waiting_times_df (pd.DataFrame): DataFrame with DEB_TIME column
    weather_df (pd.DataFrame): DataFrame with dt_iso and weather metrics

    Returns:
    pd.DataFrame: waiting_times_df with added weather columns
    """
    # Make a copy to avoid modifying original
    result_df = waiting_times_df.copy()

    # Convert waiting times to datetime if needed
    if not pd.api.types.is_datetime64_any_dtype(result_df["DEB_TIME"]):
        result_df["DEB_TIME"] = pd.to_datetime(result_df["DEB_TIME"])

    # Clean and convert weather timestamps
    # Remove the "+0000 UTC" part and convert to datetime
    weather_df = weather_df.copy()
    weather_df["dt_iso_clean"] = weather_df["dt_iso"].str.replace(" +0000 UTC", "")
    weather_df["dt_iso_clean"] = pd.to_datetime(weather_df["dt_iso_clean"])

    # Create arrays of timestamps for efficient searching
    wait_times = result_df["DEB_TIME"].values.astype("datetime64[s]")
    weather_times = weather_df["dt_iso_clean"].values.astype("datetime64[s]")

    # Find indices of nearest weather timestamps for each waiting time
    nearest_idx = []
    for t in wait_times:
        # Convert to int64 for numerical comparison
        time_diffs = abs(weather_times.astype("int64") - t.astype("int64"))
        nearest_idx.append(time_diffs.argmin())

    # Add weather data columns to result
    weather_columns = ["dew_point", "feels_like", "rain_1h", "snow_1h", "clouds_all"]
    for col in weather_columns:
        result_df[col] = weather_df[col].values[nearest_idx]

    # Add the matched timestamp for reference
    result_df["weather_timestamp"] = weather_df["dt_iso_clean"].values[nearest_idx]

    return result_df
```

#### 1.6.4 Waiting with Closing (Entity)

```python
def process_closures(waiting_times, entity_schedule, chunk_size=100000):
    """
    Memory-efficient version for processing attraction closures
    """
    # Convert dates once at the start if needed
    if not pd.api.types.is_datetime64_any_dtype(waiting_times["DEB_TIME"]):
        waiting_times["DEB_TIME"] = pd.to_datetime(waiting_times["DEB_TIME"])
    if not pd.api.types.is_datetime64_any_dtype(entity_schedule["DEB_TIME"]):
        entity_schedule["DEB_TIME"] = pd.to_datetime(entity_schedule["DEB_TIME"])
        entity_schedule["FIN_TIME"] = pd.to_datetime(entity_schedule["FIN_TIME"])

    # Create schedule lookup dictionary for faster processing
    schedule_dict = {}
    for _, row in entity_schedule.iterrows():
        entity = row["ENTITY_DESCRIPTION_SHORT"]
        if entity not in schedule_dict:
            schedule_dict[entity] = []
        schedule_dict[entity].append(
            {
                "deb": row["DEB_TIME"],
                "fin": row["FIN_TIME"],
                "desc": row["REF_CLOSING_DESCRIPTION"],
            }
        )

    # Initialize results with minimal memory
    result_df = waiting_times.copy()
    result_df["is_closed"] = 0
    result_df["REF_CLOSING_DESCRIPTION"] = None
    result_df["simultaneous_closures"] = 0

    # Process in chunks to avoid memory issues
    for start_idx in range(0, len(waiting_times), chunk_size):
        end_idx = min(start_idx + chunk_size, len(waiting_times))
        chunk = waiting_times.iloc[start_idx:end_idx]

        # Process closures for this chunk
        for idx, row in chunk.iterrows():
            entity = row["ENTITY_DESCRIPTION_SHORT"]
            deb_time = row["DEB_TIME"]

            if entity in schedule_dict:
                for schedule in schedule_dict[entity]:
                    if schedule["deb"] <= deb_time <= schedule["fin"]:
                        result_df.loc[idx, "is_closed"] = 1
                        result_df.loc[idx, "REF_CLOSING_DESCRIPTION"] = schedule["desc"]
                        break

    # Calculate simultaneous closures in chunks
    unique_times = result_df["DEB_TIME"].unique()
    closure_counts = {}

    for time in unique_times:
        mask = result_df["DEB_TIME"] == time
        closure_counts[time] = result_df.loc[mask, "is_closed"].sum()

    result_df["simultaneous_closures"] = result_df["DEB_TIME"].map(closure_counts)

    return result_df
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def generate_predictions(training_df, weather_df, start_date=None, months=3):
    """
    Generate predictions DataFrame with 15-minute intervals for specified period.
    Super-optimized version focusing on numpy operations.
    """
    # Initial datetime conversions - only convert if not already datetime
    training_df['DEB_TIME'] = pd.to_datetime(training_df['DEB_TIME'])
    
    # No need to convert dt_iso if it's already datetime
    if not pd.api.types.is_datetime64_any_dtype(weather_df['dt_iso']):
        weather_df['dt_iso'] = pd.to_datetime(weather_df['dt_iso'])
    
    # Set up time range
    if start_date is None:
        start_date = training_df['DEB_TIME'].max() + timedelta(minutes=15)
    end_date = start_date + pd.DateOffset(months=months)
    date_range = pd.date_range(start=start_date, end=end_date, freq='15T')
    
    # Get unique attractions
    attractions = training_df['ENTITY_DESCRIPTION_SHORT'].unique()
    
    # Pre-calculate hourly averages using numpy operations
    training_df['hour'] = training_df['DEB_TIME'].dt.hour
    agg_cols = ['NB_UNITS', 'GUEST_CARRIED', 'CAPACITY', 'ADJUST_CAPACITY', 'is_closed', 'simultaneous_closures']
    
    # Create lookup arrays for fast access
    hours = np.arange(24)
    attr_to_idx = {attr: idx for idx, attr in enumerate(attractions)}
    hour_attr_averages = np.full((len(attractions), 24, len(agg_cols)), np.nan)
    
    # Calculate hourly averages using numpy
    for attr in attractions:
        mask = training_df['ENTITY_DESCRIPTION_SHORT'] == attr
        attr_data = training_df[mask]
        for h in hours:
            hour_mask = attr_data['hour'] == h
            if hour_mask.any():
                values = attr_data[hour_mask][agg_cols].mean().values
                hour_attr_averages[attr_to_idx[attr], h] = values
    
    # Calculate overall averages for fallback
    overall_averages = np.zeros((len(attractions), len(agg_cols)))
    for attr in attractions:
        mask = training_df['ENTITY_DESCRIPTION_SHORT'] == attr
        overall_averages[attr_to_idx[attr]] = training_df[mask][agg_cols].mean().values
    
    # Prepare weather data
    weather_times = weather_df['dt_iso'].values.astype('datetime64[ns]')
    weather_data = weather_df[['dew_point', 'feels_like', 'rain_1h', 'snow_1h', 'clouds_all', 'dt_iso']].values
    
    # Create output arrays
    n_timestamps = len(date_range)
    n_attractions = len(attractions)
    total_rows = n_timestamps * n_attractions
    
    # Initialize output arrays
    output_times = np.repeat(date_range.values, n_attractions)
    output_attractions = np.tile(attractions, n_timestamps)
    output_hours = np.repeat([d.hour for d in date_range], n_attractions)
    
    # Initialize metrics arrays
    metrics = np.zeros((total_rows, len(agg_cols)))
    
    # Fill metrics using vectorized operations
    for i, attr in enumerate(attractions):
        idx_start = i
        idx_end = total_rows
        step = n_attractions
        hour_idx = output_hours[idx_start:idx_end:step]
        metrics[idx_start:idx_end:step] = hour_attr_averages[i, hour_idx]
    
    # Fill NaN values with overall averages
    for i in range(len(agg_cols)):
        nan_mask = np.isnan(metrics[:, i])
        if nan_mask.any():
            attr_idx = np.array([attr_to_idx[attr] for attr in output_attractions[nan_mask]])
            metrics[nan_mask, i] = overall_averages[attr_idx, i]
    
    # Find closest weather timestamps using vectorized operations
    output_times_arr = output_times.astype('datetime64[ns]')
    weather_idx = np.searchsorted(weather_times, output_times_arr)
    weather_idx = np.clip(weather_idx, 0, len(weather_times) - 1)
    
    # Get weather data
    closest_weather = weather_data[weather_idx]
    
    # Create final DataFrame
    result_df = pd.DataFrame({
        'WORK_DATE': pd.to_datetime(output_times).date,
        'DEB_TIME': output_times,
        'DEB_TIME_HOUR': output_hours,
        'ENTITY_DESCRIPTION_SHORT': output_attractions,
        'WAIT_TIME_MAX': np.zeros(total_rows),
        'NB_UNITS': metrics[:, 0],
        'GUEST_CARRIED': metrics[:, 1],
        'CAPACITY': metrics[:, 2],
        'ADJUST_CAPACITY': metrics[:, 3],
        'is_closed': np.round(metrics[:, 4]),
        'simultaneous_closures': np.round(metrics[:, 5]),
        'OPEN_TIME': None,
        'daily_attendance': None,
        'dew_point': closest_weather[:, 0],
        'feels_like': closest_weather[:, 1],
        'rain_1h': closest_weather[:, 2],
        'snow_1h': closest_weather[:, 3],
        'clouds_all': closest_weather[:, 4],
        'weather_timestamp': closest_weather[:, 5],
        'REF_CLOSING_DESCRIPTION': None
    })
    
    # Final sorting and cleanup
    result_df = result_df.sort_values(['DEB_TIME', 'ENTITY_DESCRIPTION_SHORT']).reset_index(drop=True)
    result_df['FIN_TIME'] = result_df['DEB_TIME'] + timedelta(minutes=15)
    
    # Ensure columns are in the correct order
    final_columns = [
        'WORK_DATE', 'DEB_TIME', 'DEB_TIME_HOUR', 'ENTITY_DESCRIPTION_SHORT',
        'WAIT_TIME_MAX', 'NB_UNITS', 'GUEST_CARRIED', 'CAPACITY', 'ADJUST_CAPACITY',
        'OPEN_TIME', 'daily_attendance', 'dew_point', 'feels_like', 'rain_1h',
        'snow_1h', 'clouds_all', 'weather_timestamp', 'is_closed',
        'REF_CLOSING_DESCRIPTION', 'simultaneous_closures', 'FIN_TIME'
    ]
    
    return result_df[final_columns]
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor
from datetime import timedelta, datetime

def create_time_features(df):
    """Create time-based features for the model."""
    df = df.copy()
    
    # Extract time features
    df['hour'] = df['DEB_TIME'].dt.hour
    df['day_of_week'] = df['DEB_TIME'].dt.dayofweek
    df['month'] = df['DEB_TIME'].dt.month
    df['day'] = df['DEB_TIME'].dt.day
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)
    
    # For training data only, create lag features
    if 'WAIT_TIME_MAX' in df.columns and df['WAIT_TIME_MAX'].notna().any():
        for lag in [1, 2, 3, 4]:  # 15min, 30min, 45min, 1hour lags
            df[f'wait_time_lag_{lag}'] = df.groupby('ENTITY_DESCRIPTION_SHORT')['WAIT_TIME_MAX'].shift(lag)
        
        # Create rolling mean features
        windows = [4, 12, 24]  # 1hour, 3hours, 6hours
        for window in windows:
            df[f'wait_time_rolling_mean_{window}'] = (
                df.groupby('ENTITY_DESCRIPTION_SHORT')['WAIT_TIME_MAX']
                .rolling(window=window, min_periods=1)
                .mean()
                .reset_index(0, drop=True)
            )
    
    return df

def train_and_evaluate_model(waiting_df, predictions_df, attraction):
    """Train model on waiting data with CV on last 3 months, then predict for future."""
    # Ensure columns match
    required_columns = [
        'WORK_DATE', 'DEB_TIME', 'DEB_TIME_HOUR', 'ENTITY_DESCRIPTION_SHORT',
        'WAIT_TIME_MAX', 'NB_UNITS', 'GUEST_CARRIED', 'CAPACITY', 'ADJUST_CAPACITY',
        'OPEN_TIME', 'daily_attendance', 'dew_point', 'feels_like', 'rain_1h',
        'snow_1h', 'clouds_all', 'weather_timestamp', 'is_closed',
        'REF_CLOSING_DESCRIPTION', 'simultaneous_closures', 'FIN_TIME'
    ]
    
    # Ensure all required columns exist in both datasets
    for df in [waiting_df, predictions_df]:
        for col in required_columns:
            if col not in df.columns:
                df[col] = None
    
    # Filter data for the attraction
    attraction_data = waiting_df[waiting_df['ENTITY_DESCRIPTION_SHORT'] == attraction].copy()
    attraction_data = attraction_data.sort_values('DEB_TIME')
    
    # Get the last 3 months for cross-validation
    last_date = attraction_data['DEB_TIME'].max()
    cv_start_date = last_date - pd.DateOffset(months=3)
    
    # Split into training and CV sets
    train_data = attraction_data[attraction_data['DEB_TIME'] <= cv_start_date]
    cv_data = attraction_data[attraction_data['DEB_TIME'] > cv_start_date]
    
    # Create features
    train_features = create_time_features(train_data)
    cv_features = create_time_features(cv_data)
    
    # Define base features (excluding lag features for predictions)
    base_features = [
        'hour', 'day_of_week', 'month', 'day', 'is_weekend',
        'NB_UNITS', 'GUEST_CARRIED', 'CAPACITY', 'ADJUST_CAPACITY',
        'dew_point', 'feels_like', 'clouds_all', 'is_closed',
        'simultaneous_closures'
    ]
    
    # Add lag features for training/CV
    lag_features = [col for col in train_features.columns if 'wait_time' in col]
    all_features = base_features + lag_features
    
    # Remove rows with NaN values
    train_features = train_features.dropna(subset=all_features + ['WAIT_TIME_MAX'])
    cv_features = cv_features.dropna(subset=all_features + ['WAIT_TIME_MAX'])
    
    # Initialize model
    model = XGBRegressor(
        n_estimators=1000,
        learning_rate=0.05,
        max_depth=6,
        min_child_weight=1,
        random_state=42
    )
    
    # Train on full training set with early stopping
    model.fit(
        train_features[all_features],
        train_features['WAIT_TIME_MAX'],
        eval_set=[(cv_features[all_features], cv_features['WAIT_TIME_MAX'])],
        early_stopping_rounds=50,
        verbose=False
    )
    
    # Perform cross-validation on the last 3 months
    cv = TimeSeriesSplit(n_splits=3)
    cv_scores = []
    
    for train_idx, val_idx in cv.split(cv_features):
        cv_train = cv_features.iloc[train_idx]
        cv_val = cv_features.iloc[val_idx]
        
        cv_model = XGBRegressor(
            n_estimators=model.best_iteration,
            learning_rate=0.05,
            max_depth=6,
            min_child_weight=1,
            random_state=42
        )
        
        cv_model.fit(cv_train[all_features], cv_train['WAIT_TIME_MAX'])
        predictions = cv_model.predict(cv_val[all_features])
        
        mae = mean_absolute_error(cv_val['WAIT_TIME_MAX'], predictions)
        rmse = np.sqrt(mean_squared_error(cv_val['WAIT_TIME_MAX'], predictions))
        cv_scores.append({'MAE': mae, 'RMSE': rmse})
    
    # Retrain model on entire dataset before making predictions
    full_features = create_time_features(attraction_data)
    full_features = full_features.dropna(subset=all_features + ['WAIT_TIME_MAX'])
    
    final_model = XGBRegressor(
        n_estimators=model.best_iteration,
        learning_rate=0.05,
        max_depth=6,
        min_child_weight=1,
        random_state=42
    )
    
    final_model.fit(full_features[all_features], full_features['WAIT_TIME_MAX'])
    
    # Prepare prediction data
    pred_data = predictions_df[predictions_df['ENTITY_DESCRIPTION_SHORT'] == attraction].copy()
    pred_features = create_time_features(pred_data)
    
    # Make predictions for future data
    try:
        predictions = final_model.predict(pred_features[base_features])
        return {
            'cv_scores': cv_scores,
            'predictions': predictions,
            'feature_importance': dict(zip(all_features, final_model.feature_importances_))
        }
    except Exception as e:
        print(f"Error predicting for {attraction}: {str(e)}")
        return None

def evaluate_and_predict(waiting_df, predictions_df):
    """Process all attractions and return combined results."""
    # Ensure both dataframes have the same columns in the same order
    required_columns = [
        'WORK_DATE', 'DEB_TIME', 'DEB_TIME_HOUR', 'ENTITY_DESCRIPTION_SHORT',
        'WAIT_TIME_MAX', 'NB_UNITS', 'GUEST_CARRIED', 'CAPACITY', 'ADJUST_CAPACITY',
        'OPEN_TIME', 'daily_attendance', 'dew_point', 'feels_like', 'rain_1h',
        'snow_1h', 'clouds_all', 'weather_timestamp', 'is_closed',
        'REF_CLOSING_DESCRIPTION', 'simultaneous_closures', 'FIN_TIME'
    ]
    
    predictions_df = predictions_df.reindex(columns=required_columns)
    waiting_df = waiting_df.reindex(columns=required_columns)
    
    results = {}
    
    for attraction in waiting_df['ENTITY_DESCRIPTION_SHORT'].unique():
        print(f"\nProcessing {attraction}:")
        result = train_and_evaluate_model(waiting_df, predictions_df, attraction)
        
        if result is not None:
            # Print CV scores
            cv_scores = result['cv_scores']
            mean_mae = np.mean([score['MAE'] for score in cv_scores])
            mean_rmse = np.mean([score['RMSE'] for score in cv_scores])
            print(f"Cross-validation scores (last 3 months):")
            print(f"Mean MAE: {mean_mae:.2f}")
            print(f"Mean RMSE: {mean_rmse:.2f}")
            
            # Print top features
            importance = result['feature_importance']
            top_features = dict(sorted(importance.items(), key=lambda x: x[1], reverse=True)[:5])
            print("\nTop 5 important features:")
            for feat, imp in top_features.items():
                print(f"{feat}: {imp:.4f}")
            
            # Store predictions
            mask = predictions_df['ENTITY_DESCRIPTION_SHORT'] == attraction
            predictions_df.loc[mask, 'WAIT_TIME_MAX'] = result['predictions']
            
            results[attraction] = result
    
    return predictions_df[required_columns], results
```
